```python
import os
import re
import chardet
from collections import Counter

# Сургалтын өгөгдлийн зам
folder_path_ham = "data/train/ham"
folder_path_spam = "data/train/spam"

# Унших файлын хязгаар
limit = 2
# Уншсан файлын тоо
count = 0

def createUnigram(folder_path):
    unigram = Counter()

    # ---------------- Файл унших хэсэг ----------------
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            # print(f"Reading: {filename}")

            with open(file_path, "rb") as f:
                raw_data = f.read()

            result = chardet.detect(raw_data)
            encoding = result["encoding"] or "utf-8"

            try:
                text = raw_data.decode(encoding, errors="ignore")
            except Exception as e:
                print(f"⚠️ Could not decode {filename}: {e}")
                continue

            cleaned = re.sub(r'[^a-zA-Z가-힣\s]', '', text).lower()
            words = cleaned.split()


            unigram.update(words)

    return unigram

```

```python
def save_unigram(unigram, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for word, count in unigram.most_common():
            f.write(f"{word} {count}\n")
    print(f"✅ Saved {filename}")
```

```python
ham_unigram = createUnigram(folder_path_ham)
spam_unigram = createUnigram(folder_path_spam)

# print("=== HAM unigram ===")
# for word, freq in ham_unigram.most_common(10):
#     print(f"{word}: {freq}")

# print("\n=== SPAM bag of words ===")
# for word, freq in spam_unigram.most_common(10):
#     print(f"{word}: {freq}")

print(f"\nTotal unique ham words: {len(ham_unigram)}")
print(f"Total unique spam words: {len(spam_unigram)}")

save_unigram(ham_unigram, "ham_unigram.txt")
save_unigram(spam_unigram, "spam_unigram.txt")
```
